```python
# Set the environment variable in the Colab session
import os

GROQ_API_KEY= os.getenv("GROQ_API_KEY")
```

```python
from crewai import LLM, Crew

# LLM - CREWAI wrapper that integrates with litellm 
groq_llm = LLM("groq",
                api_key=GROQ_API_KEY,
               model = "groq/llama-3.1-8b-instruct", 
               temperature=0.7)


```

role - defines identity
goal - defines objective 
backstory - shapes resoning
llm - powers the agent
verbose - shows internal execution laws

```python
from crewai import Agent

researcher = Agent(role = "Researcher Analyst", 
                   goal = "Conduct structured research on a topic",
                   backstory = "Expert in gathering accurate AI insights. ",
                #    constraints = "Use credible sources for research Organize findings in a structured format",
                   llm=groq_llm, verbose=True)

writer = Agent(role = "Technical Writer", 
               goal = "Write a detailed and clear blog post",
               backstory = "Skilled at explaining AI systems clearly",
               llm=groq_llm, verbose=True)

editor = Agent(role = "Content Editor", 
               goal = "Refine and polish the content professionally",   
               backstory = "Improves clarity , grammer and structure.",
               llm=groq_llm, verbose=True)
```

Why expected output matters?
- It forces structure
- removes randomness
- Improves reliability

```python
# Create tasks for each agent - exccution layer of the crew

from crewai import Task

research_task = Task(
    name="Research Task", 
    description="Gather detailed information on a specific topic.",
    expected_output="Structured research notes ....",
    agent=researcher)

writing_task = Task(
    name="Writing Task", 
    description="Write a comprehensive blog post based on the research.", 
    expected_output="A well-written blog post draft.",
    agent=writer)

editing_task = Task(
    name="Editing Task", 
    description="Edit and refine the blog post for clarity and professionalism.", 
    expected_output="A polished and professional blog post.",
    agent=editor)
```

Orchestration - Create the Crew

Process.sequential means
Research -> writing -> editing

Each task feeds into the next one

```python
from crewai import Process

crew = Crew(name="Content Creation Crew",
            agents=[researcher, writer, editor],
            tasks=[research_task, writing_task, editing_task],  
            process = Process.sequential,
            verbose=True)

        
```

Execute the system
- Research agent runs
- Writer consumes research
- Editor refines the final draft

```python
result = crew.kickoff()
print("Final Output:", result)
```
