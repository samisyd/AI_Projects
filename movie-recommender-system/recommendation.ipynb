```python
import ast
```

```python
# this will convert the string representation of list of dictionaries into a list of genres 
ast.literal_eval(movies.iloc[0].genres)
```

```python
# helper function to convert the string representation of list of dictionaries into a list of genres
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies.iloc[0].cast
```

```python
# for cast we will keep only top 3 actors, and for crew we will keep only director, all important actors are in the 
# beginning of the list, and for crew we will keep only director as they are the most important for the movie 
# and also they are the most important for the recommendation system
# charcter is not needed , we need only actor name and director name for our recommendation system
def convert2(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert2)
movies.head()
# so we are getting only 3 actors
```

```python

# where job is director and we need only that dictonary. we dont need any other dict.
# we need to extract the director name from the crew column and create a list of directors for each movie, as there can be multiple directors for a movie
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python

# now we will concatenate all the tags into a single list and convert to a single string for each movie so thats the tag the final column#
# we need to do a transformation on the directorname/ cast - so the recommendation system doesnt get confused between 
# "sam medola" and "sam august" so we concatenate it to sammedola and samaugust
# can be done this way
# movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
# movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
# movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
# movies.head() 
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
new['tags'][1]
# find similarity score between tags. find same words in the tags 
# but we need to convert into vectors 
# we will pick the closest vectors , the closest neighbor in the 2 D vector space 
# will be the most similar movie to the movie we are looking for

# TECHNIQUES - BAG OF WORDS, TF IDF, WORD EMBEDDINGS
#BAG OF WORDS
# so combine all tags and get top 5000 words whose frequency is highest WE WILL EXTRACT THAT.
# WORD1 WORD2 WORD3 WORD4 WORD5

#TAG1 + TAG2 + TAG3 + TAG4 + TAG5
# LARGE TEXT DATASET - 5000 MOST COMMON WORDS IN THIS DATASET - 5000 DIMENSIONAL VECTOR SPACE
# FOR EACH tag we will create a vector of 5000 dimensions and we will put 1 or the occurance of that word  is 
# present in the tag and 0 if the word is not present in the tag

#    w1  w2 w3 w4 w5 .... w5000
# t1  5   3  0  0  1
# t2  2   1  0  1  0
# t3  1   1  0  0  1
# t4  5   2  5  1  0

# so if we pick the 1st 2 colums for ex it will be a 2D space and where they points lie in the space will determine the 
# similarity between the tags, if they are close to each other then they are similar and if they are far from each 
# other then they are not similar

# also we will remove stop words like "the","a","an" etc as they are not important for our recommendation system 
# and they will not help us in finding the similarity between the movies

# we will use skilearn's CountVectorizer to convert the tags into vectors and then we will calculate the 
# similarity between the movies based on the vectors and then we will recommend the movies based on the similarity score



```

```python
# now do text vectorization, we will use count vectorizer to convert the tags into vectors and then we will calculate the
#  similarity between the movies based on the vectors and then we will recommend the movies based on the similarity score

# need to develop a website where user can input the movie name and then we will recommend the movies based on the similarity score 
# similar 5 movies based on the tags

from sklearn.feature_extraction.text import CountVectorizer
# remove stop words like "the","a","an" etc as they are not important for our recommendation system
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer  
ps = PorterStemmer()
print(ps.stem('loved'))
print(ps.stem('loving'))
# u will get love for both the words after stemming.
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

# shows an ex of what stem does to the tags
stem(new['tags'][0])
```

```python
new['tags'] = new['tags'].apply(stem)

#  now run again the count vectorizer to convert the tags into vectors and then we will calculate the
# remove stop words like "the","a","an" etc as they are not important for our recommendation system
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new['tags']).toarray()
# now every moview is in avector form
print(vector.shape)
vector[0] 
```

```python
cv.get_feature_names_out()
feature_names = cv.get_feature_names_out().tolist()
print(feature_names)
#  all the words are displayed here, we can see that the most common words are "love
#now words are not repeating
```

```python
#  now every move is a vector - 4806 movies  and each movie has 5000 numbers
# every movie with other movie find similarity 
# we will calculate cosine similarity - and not euclidian space
# the angle between 2 vestors , the less the angle/ theta the similar 
# theta between 2 vectors is less the movies are similar - find cosine distance
# use cosine similarity from sklearn

from sklearn.metrics.pairwise import cosine_similarity
```

```python
# find with every vector the distance with every other vector
# every movies distance with every other movie  - 1 movie compared to 4086 movies then 2nd movie same and so on..
# 4806* 4806 - storing in similarity matrix - array of array 
print(cosine_similarity(vector).shape)
similarity = cosine_similarity(vector)
```
